```python
cars = pd.read_csv('./project_clean.csv')
cars
```

```python
cars['isD4D'] = (cars.Model.str.upper().str.find('D4D') >= 0)*1
cars['isTERRA'] = (cars.Model.str.upper().str.find('TERRA') >= 0)*1
cars['isSOL'] = (cars.Model.str.upper().str.find('SOL') >= 0)*1
cars['isVVT'] = (cars.Model.str.upper().str.find('VVT') >= 0)*1
cars['isSPORT'] = (cars.Model.str.upper().str.find('SPORT') >= 0)*1
cars['isVVTLI'] = (cars.Model.str.upper().str.find('VVTLI') >= 0)*1
cars['isCOMFORT'] = (cars.Model.str.upper().str.find('COMF') >= 0)*1
cars['isLINEA'] = (cars.Model.str.upper().str.find('LINEA') >= 0)*1
cars['isLUNA'] = (cars.Model.str.upper().str.find('LUNA') >= 0)*1
cars['isWAGON'] = ((cars.Model.str.upper().str.find('WAGON') >= 0) | (cars.Model.str.upper().str.find('WAGEN') >= 0))*1
cars['isLIFT'] = (cars.Model.str.upper().str.find('LIFT') >= 0)*1
cars['isHATCH'] = (cars.Model.str.upper().str.find('HATCH') >= 0)*1
cars['isSEDAN'] = (cars.Model.str.upper().str.find('SEDAN') >= 0)*1
cars['isVERSO'] = (cars.Model.str.upper().str.find('VERSO') >= 0)*1
```
